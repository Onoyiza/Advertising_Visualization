```python
# Importing pandas and numpy
import pandas as pd
import numpy as np

# Reading the csv file into a pandas DataFrame
data = pd.read_csv('advertising (1).csv')

# Printing out the first 10 rows of our data
data.head(10)
```

**Plot showing how daily time spent on site and daily internet usage of the copy writer affects whether a user clicked on the ad on not**

```python
# Importing matplotlib
import matplotlib.pyplot as plt

# Function to help us plot
def plot_points(data):
    X = np.array(data[['Daily Time Spent on Site',"Daily Internet Usage"]])
    y = np.array(data["Clicked on Ad"])
    admitted = X[np.argwhere(y==1)]
    rejected = X[np.argwhere(y==0)]
    plt.scatter([s[0][0] for s in rejected], [s[0][1] for s in rejected], s = 25, color = 'red', edgecolor = 'k', label='Clicked on Ad=0')
    plt.scatter([s[0][0] for s in admitted], [s[0][1] for s in admitted], s = 25, color = 'cyan', edgecolor = 'k', label='Clicked on Ad=1')
    plt.xlabel('Daily Time Spent on Site')
    plt.ylabel('Daily Internet Usage')
    plt.title('Advertising_Data')
    
# Plotting the points
plot_points(data)
plt.legend()
plt.show()
```
